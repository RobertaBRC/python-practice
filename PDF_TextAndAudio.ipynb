### Extracting text and audio from a PDF file

```python
# Importing the required libraries
import pyttsx3
import pdfplumber
import PyPDF2
```

```python
# Opening the PDF file and creating an object from it
pdf_object = PyPDF2.PdfFileReader('sample.pdf')
```

```python
# Finding the number of pages in the document
pages = pdf_object.numPages
pages
```

```python
# Logic to get the content of the first page
# page_object = pdf_object.getPage(0)
# page_content = page_object.extractText()
```

```python
# Creating a function using the rationale above to extract the text from all pages
def getText(num_pages):
    all_pages = ''
    for i in range(0, num_pages):
        page_object = pdf_object.getPage(i)
        all_pages += page_object.extractText()
    return all_pages 

content = getText(pages)
content
```

---

Saving the data to a txt file:

```python
with open('pdf_text_output.txt', 'w') as f:
    f.write(content)
    
```

---

Extracting audio from the same file:

```python
# Playing the audio generated from the PDF
audio = pyttsx3.init()
audio.say(content)
audio.runAndWait()
audio.stop()
```

```python
# Saving it as a mp3 file
audio.save_to_file(content, 'pdf_audio_output.mp3')
audio.runAndWait()
audio.stop()
```
