_Change case according preceding letter:_
- _If preceding letter appears before the current letter in the alphabet, make the current letter upper case_
- _If preceding letter appears after the current letter in the alphabet, make the current letter lower case_

```python
def transformSentence(sentence):
    seq = 'abcdefghijklmnopqrstuvwxyz'
 
    for pos, i in enumerate(sentence):
        if i.lower() in seq: 
            if pos == 0:
                print(i, end = '')
            elif sentence[pos - 1] == ' ':
                print()
                print(i, end = '')
            else:
                if seq.index(i.lower()) < seq.index(sentence[pos-1].lower()):
                    print(i.lower(), end = '')
                elif seq.index(i.lower()) > seq.index(sentence[pos-1].lower()):
                    print(i.upper(), end = '')
                else:
                    print(i)
    
# Using the function
sentence = input()
result = transformSentence(sentence)
```
