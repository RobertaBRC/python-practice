_Rock - Paper - Scissors (JO - KEN - PO)_

```python
from random import randint 
from time import sleep

items = ('ROCK', 'PAPER', 'SCISSORS')
input_player = int(input('Choose an option: \n[0] Rock \n[1] Paper \n[2] Scissors'))
input_computer = randint(0,2)

if input_player in (0,1,2):
    if input_player == input_computer:
        res = 'DRAW'
        
    elif input_player == 0 and input_computer == 1:
        res = 'You LOST'
        print()
    elif input_player == 0 and input_computer == 2:
        res = 'You WON'
        
    elif input_player == 1 and input_computer == 0:
        res = 'You WON'
    elif input_player == 1 and input_computer == 2:
        res = 'You LOST'
        
    elif input_player == 2 and input_computer == 0:
        res = 'You LOST'
    elif input_player == 2 and input_computer == 1:
        res = 'You WON'
        
    print('JO')
    sleep(1)
    
    print('KEN')
    sleep(1)
    
    print('PO!!!')
    sleep(1)
    
    print(30 * '=')
    print(f'You played {items[input_player]} \nThe computer played {items[input_computer]}')
    print(30 * '=')
    print(res)

else:
    (print('Invalid choice. Try again'))
```
