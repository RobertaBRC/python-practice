_Sorting an array __a__ using the Bubble Sort algorithm:_

```python
a = list(map(int, input().rstrip().split()))
b = sorted(a)
count = 0
while a != b:
    for pos, i in enumerate(a):
            if pos < len(a) - 1 and i > a[pos + 1]:
                    a[pos] = a[pos + 1]
                    a[pos + 1] = i
                    count += 1
print(f'Array is sorted in {count} swaps.')
print(f'First element: {a[0]}')
print(f'Last element: {a[-1]}')
```
