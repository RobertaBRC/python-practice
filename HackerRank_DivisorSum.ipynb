_Function to calculate the sum of all divisors of a given number:_

```python
def divisorSum(n):
        divisors = []
        for i in range(n, 0, -1):
            if n % i == 0:
                divisors.append(i)
        return(sum(divisors))

# Testing the newly created function:
divisorSum(6)

# SAVED TO GITHUB
```
